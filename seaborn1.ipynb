## Hands on! 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
sales = pd.read_csv('data/sales_data.csv',parse_dates=['Date'])
```

```python
sales.head(2)
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)


SCATTERPLOT

```python
sns.scatterplot(x='Order_Quantity', y='Profit', data = sales)
plt.show()
```

```python
sns.scatterplot(x = 'Order_Quantity',y = 'Profit', hue = 'Customer_Gender', data = sales)
plt.show()
```

```python
sns.scatterplot(data = sales, x = 'Order_Quantity',y = 'Profit', hue = 'Customer_Gender', palette = {'M':'black', 'F':'red'})
plt.show()
```

```python
sns.scatterplot(data = sales, x  = 'Order_Quantity', y = 'Profit',hue = 'Customer_Gender', hue_order= ['F', 'M'])
plt.show()
```

countplot

```python
sns.countplot(data = sales, x = 'Age_Group')
plt.show()
```

```python
ax = sns.countplot(data =sales, x = 'Age_Group', hue='Customer_Gender')
plt.show()
```

```python
sns.countplot(data = sales, x = 'Age_Group',hue = 'Customer_Gender',hue_order = ['F','M'], palette= { 'F': 'red', 'M':'green'} )
plt.show()
```

RELPLOT


```python
sns.relplot(kind = 'scatter',data = sales, x =  'Order_Quantity', y= 'Profit')
plt.show()
```

```python
sns.relplot(kind = 'scatter', data = sales, x = 'Order_Quantity', y = 'Profit', col = 'Customer_Gender')
plt.show()
```

```python
sns.relplot(kind = 'scatter', data = sales, x = 'Order_Quantity', y = 'Profit', row = 'Customer_Gender', style = 'Customer_Gender')
plt.show()
```

```python
sns.relplot(kind = 'line', data = sales, x = 'Order_Quantity', y = 'Profit', hue = 'Customer_Gender', style='Customer_Gender')
plt.show()
```

```python
sns.relplot(kind = 'line', data = sales, x = 'Order_Quantity', y = 'Profit')
plt.show()
```

```python
sns.relplot(kind = 'line', data = sales, x = 'Order_Quantity', y = 'Profit',ci = 'sd')
plt.show()
```
